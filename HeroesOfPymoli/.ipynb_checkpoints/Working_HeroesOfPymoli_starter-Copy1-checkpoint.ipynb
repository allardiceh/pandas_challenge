## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
##figure out max min
#print(purchase_data["Age"].max())
#print(purchase_data["Age"].min())

# Create bins in which to place values based upon Age
bins = [ 0,9, 14, 19, 24, 29,
        34, 39, 45]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

purchase_data["Age Count"]= pd.cut(purchase_data["Age"], bins, labels=group_labels)

##use group by to get the df
Age_sort=purchase_data.groupby(by='Age Count')
Total_Count=Age_sort["SN"].nunique()
Total_sum=Total_Count.sum()
Percent_Players =(Total_Count/Total_sum)*100



Age_df = pd.DataFrame({"Total Players": Total_Count,
                          "Percent of Players":Percent_Players
                                      })

Age_df.head(8)
```

```python
purchase_sn_data=purchase_data.groupby(by='SN')

purch_count=purchase_sn_data["Item ID"].count()

purch_mean=purchase_sn_data["Price"].mean()



purchase_sum=purchase_sn_data["Price"].sum()


purch_summary=pd.DataFrame({"Total Items Count":purch_count ,
                            "Average Purchase Price":purch_mean,
                            "Total Revenue" :purchase_sum })

purch_summary.sort_values(by=['Total Revenue'], inplace=True , ascending=False)

purch_summary.head()
```
